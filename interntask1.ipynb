```python
# convert "Start Date" column to datetime data type
df['Start Date'] = pd.to_datetime(df['Start Date'], format='%m/%d/%Y', errors='coerce')
```

```python
df.nunique()
```

```python
df.isnull().sum()
```

```python
df["Gender"].fillna("No Gender", inplace = True)

df.isnull().sum()
```

```python
mode = df['Senior Management'].mode().values[0]
df['Senior Management']= df['Senior Management'].replace(np.nan, mode)

df.isnull().sum()
```

```python
average_salary = df['Salary'].mean()
print("Average Salary:", average_salary)
```

```python

grouped_df = df.groupby(['Salary'])['Team'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None)
print(grouped_df)
```
